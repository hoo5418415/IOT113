```python
import pandas as pd
from pycaret.classification import setup, compare_models

# Load the datasets
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

# Data cleaning and feature engineering
# For simplicity, handle missing values, convert categorical variables, and drop unnecessary columns
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data['Fare'].fillna(train_data['Fare'].median(), inplace=True)

# Convert categorical variables to numeric
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)

# Drop columns that are not helpful
train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Define target and features
target = 'Survived'

# Setup PyCaret
clf_setup = setup(data=train_data, target=target, verbose=False, session_id=123)

# Compare models
best_model = compare_models()

# Save the PyCaret environment
#save_config("pycaret_titanic_env.pkl")

# Output the best model
print("Best Model: ", best_model)
```
